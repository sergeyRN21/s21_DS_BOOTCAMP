```python
import pandas as pd
```

<h2>Загрузите файл JSON, который вы создали в предыдущем упражнении, во фрейм данных</h2>

```python
df = pd.read_json('../data/auto.json', orient = 'records')
df.set_index('CarNumber', inplace=True)
df
```

<h2>Выполните следующие действия:</h2>

```python
#отображайте только те строки, в которых штрафы превышают 2100
df[df['Fines']> 2100]
```

```python
#отображать только те строки, в которых штрафы превышают 2100, а возврат средств равен 2
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
#отображать строки только в тех случаях, когда модели указаны в списке: [’Focus’, ’Corolla’]
df[df['Model'].isin(['Focus', 'Corolla'])]
```

```python
#отображать только те строки, в которых указан номер автомобиля из списка:
#[’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]
nums = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(nums)]
```

<h2>Cоздайте агрегированные данные с указанием марки и модели</h2>

```python
#отоброжать средные штрафы сгрупированные по марке авто
df.groupby(by=['Mark']).agg({'Fines':['median']})
```

```python
#отобразите средние значения штрафов, сгруппированные по марке и модели
df.groupby(by=['Mark', 'Model']).agg({'Fines':['median']})
```

```python
#отобразите количество штрафов, сгруппированных по маркам и моделям, чтобы понять, можно ли доверять средним значениям
df.groupby(by=['Mark', 'Model']).agg({'Fines':['count']})
```

```python
#отобразите минимальные и максимальные штрафы, сгруппированные по маркам и моделям, чтобы лучше понять разницу
df.groupby(by=['Mark', 'Model']).agg({'Fines':['min', 'max']})
```

```python
#отобразите стандартное отклонение штрафов, сгруппированных по маркам и моделям, чтобы лучше понять разброс
df.groupby(by=['Mark','Model']).agg({'Fines': ['std']})
```

<h2>Произведите агрегацию с указанием номера автомобиля</h2>

```python
# отобразить номера автомобилей, сгруппированные по количеству штрафов в порядке убывания,
# мы хотим найти тех, кто чаще всего нарушал закон
df.groupby(by=['CarNumber']).agg({'Fines':['count']}).sort_values(by=('Fines','count'), ascending=False)
```

```python
#выберите из исходного фрейма данных все строки, соответствующие номеру автомобиля, который находится на первом месте, 
#мы хотим немного увеличить масштаб
df[df.index == 'Y7689C197RUS']
```

```python
#выведите номера автомобилей, сгруппированные по сумме штрафов в порядке убывания, мы хотим найти тех, кто заплатил больше всего
df.groupby(by=['CarNumber']).agg({'Fines': ['sum']}).sort_values(by=('Fines', 'sum'), ascending=False)
```

```python
#выберите из исходного фрейма данных все строки, соответствующие номеру автомобиля, 
#который находится на первом месте, мы хотим немного увеличить масштаб
df[df.index == 'X758HY197RUS']
```

```python
#выведите таблицу, которая отвечает на вопрос: есть ли номера автомобилей, которые были связаны с разными моделями?
df.groupby(by=['CarNumber', 'Model']).agg({'Model': ['nunique']}).sort_values(by=('Model', 'nunique'), ascending=False)
```
