```python
import pandas as pd
```

<h2> Создание фрейма данных с двумя столбцами и преобразование datetime в Dtype datetime64[ns]</h2>

```python
df = pd.read_csv('../data/feed-views.log', names = ['datetime', 'user'], parse_dates = ['datetime'], sep = '\t')
```

```python
df.info()
```

<h2>Извлечение года, месяца, дня, часа, минуты секунды в другой столбец</h2>

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
df
```

<h2>Создание нового столбца daytime</h2>
создайте новый столбец daytime
вам нужно присвоить конкретное значение времени суток, если час находится в определённом интервале, например, если час больше 11 и меньше или равен 17.

- 0 — 3.59 ночь, 4 — 6.59 раннее утро, 7 — 10.59 утро, 11 — 16.59 день, 17 — 19.59 раннее вечернее время, 20 — 23.59 вечернее время
- назначить user в качестве индекса

```python
bin = [-1, 3, 6, 10, 16, 19, 23]
label = ['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
df['daytime'] = pd.cut(df['hour'], bins=bin,labels=label)
df.set_index('user', inplace=True)
df
```

<h2>Рассчитайте минимум и максимум для часов и режима для категорий дневного времени</h2>

1. максимальное количество часов для строк, где время суток равно ночи

```python
maxvalue = df[df['daytime'] == 'night']
maxvalue = maxvalue['hour'].max()
maxvalue
```

2. минимальное количество часов для строк, где время суток - утро

```python
minvalue = df[df['daytime'] == 'morning']
minvalue = minvalue['hour'].min()
minvalue
```
