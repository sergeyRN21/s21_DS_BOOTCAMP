```python
import pandas as pd
```

<h2>Прочитайте CSV-файл и сделайте ID индексным столбцом</h2>

```python
df = pd.read_csv('../data/auto.csv', index_col = 'ID')
df
```

<h2>Подсчитайте количество наблюдений с помощью метода count()</h2>

```python
df.count()
```

<h2>отбросить дубликаты, учитывая только следующие столбцы: CarNumber, Make_n_model, Fines</h2>

- между двумя равными наблюдениями вам нужно выбрать последнее

- проверьте количество наблюдений еще раз

```python
df.drop_duplicates(subset= ['CarNumber', 'Make_n_model', 'Fines'], keep='last',inplace=True)
df
```

<h2>Работа с пропущенными значениями</h2>

```python
# проверьте, сколько значений отсутствует в каждом столбце
df.isnull().sum()
```

```python
# удалите все столбцы с более чем 500 пропущенными значениями,
# используя аргумент thresh, проверьте, сколько пропущенных значений в каждом столбце
df.dropna(axis='columns',
         thresh=500,
         inplace=True)
df.isnull().sum()
```

```python
# замените все пропущенные значения в столбце `Refund` предыдущим значением в этом столбце
# для этой ячейки, используйте метод аргумента, проверьте, сколько значений отсутствует в каждом столбце
df['Refund'] = df['Refund'].ffill()
df.isnull().sum()
```

```python
# замените все пропущенные значения в столбце «Fines» средним значением этого столбца
#(исключите значения NA/null при вычислении среднего значения),
# проверьте, сколько значений отсутствует в каждом столбце
mean_fines = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(value=mean_fines)
df.isnull().sum()
```

<h2>Разделить и проанализировать марку и модель</h2>

```python
# используйте метод apply как для разделения, так и для извлечения значений в новые столбцы Make и Model.
df['Mark'] = df['Make_n_model'].apply(lambda x:x.split()[0])
df['Model'] = df['Make_n_model'].apply(lambda x:x.split()[1] if len(x.split()) == 2 else None)

# отбросьте столбец Make_n_model
df.drop('Make_n_model', axis='columns', inplace=True)
df
```

```python
df.to_json('../data/auto.json', orient='records', double_precision=1)
```
