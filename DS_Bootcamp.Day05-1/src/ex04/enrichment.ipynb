<h2>Обогатите фрейм данных, используя образец из этого фрейма данных</h2>

создайте выборку из 200 новых наблюдений со случайным_состоянием = 21
- в выборке не должно быть новых комбинаций номера автомобиля, марки и модели, поэтому весь набор данных будет согласован по этим параметрам
- нет ограничений на возврат средств и штрафы, вы можете взять любое значение из этих столбцов наугад и использовать его для любого номера автомобиля

```python
sample = df.sample(n=200, random_state=21)
sample
```

```python
uniqueRefund = df['Refund'].unique()
uniqueRefund
```

```python
uniqueFines = df['Fines'].unique()
```

```python
sample['Refund'] = np.random.choice(uniqueRefund, size=len(sample))
sample['Fines'] = np.random.choice(uniqueFines, size=len(sample))
sample
```

<h2>Объедините выборку с исходным фреймом данных в новый фрейм данных con- cat_rows</h2>

```python
concat_rows = pd.concat(objs=[df,sample], ignore_index=True)
concat_rows
```

<h2>Обогатите фрейм данных concat_rows новым столбцом с сгенерированными данными</h2>

- создайте серию с названием «Год», используя случайные целые числа от 1980 до 2019
- используйте np.random.seed(21) перед генерацией лет
- объедините серию с фреймом данных и назовите ее fines

```python
np.random.seed(21)

Year = pd.Series([np.random.randint(1980,2019) for i in range(len(concat_rows))], name = 'Year')
Year
```

```python
fines = pd.concat([concat_rows, Year], axis='columns')
fines
```

<h2>Обогатите фрейм данных данными из другого фрейма данных</h2>

<h3>создайте новый фрейм данных с номерами автомобилей и их владельцами</h3>
- получите самые популярные фамилии (вы можете найти файл surname.json во вложениях) в США
- создайте новую серию с фамилиями (в них не должно быть специальных символов, таких как запятые, скобки и т. д.) на основе собранных вами данных.Количество должно быть равно числу уникальных номеров автомобилей в выборке (используйте random_state = 21)
- создайте владельцев фрейма данных с 2 столбцами: номер и ФАМИЛИЯ

```python
surname = pd.read_json('../data/surname.json', orient='values')
surname.columns=surname.iloc[0]
surname.drop(0,inplace=True)
surname
```

```python
car_numbers = concat_rows.drop_duplicates('CarNumber')['CarNumber']
car_numbers = car_numbers.to_frame(name='CarNumber').reset_index(drop=True)
car_numbers
```

```python
surnames = surname['NAME'].sample(n=len(car_numbers), random_state=21,replace=True)
surnames = surnames.to_frame(name='SURNAME').reset_index(drop=True)
surnames
```

```python
owners = pd.concat([car_numbers, surnames],axis=1, ignore_index=False)
owners
```

<h2>Удалите последние 20 наблюдений из фрейма данных owners и добавьте 3 новых наблюдения (они не совпадают с теми, которые вы добавляете во фрейм данных штрафов)</h2>

```python
owners.drop(owners.tail(20).index, inplace=True)
owners
```

```python
new_surname = [['C111CC66RUS', 'IVANOV'],
            ['C222CC66RUS', 'VASILYEV'],
            ['C333CC88RUS', 'ISAEV']]
new_surname = pd.DataFrame(new_surname, columns=owners.columns)
owners = pd.concat([owners, new_surname])

owners
```

<h2>Объединить оба фрейма данных:</h2>

```python
#новый фрейм данных должен содержать только номера автомобилей, которые существуют в обоих фреймах данных
fines_1 = pd.merge(left=fines, right=owners, on='CarNumber', how='inner')
fines_1
```

```python
#новый фрейм данных должен содержать все автомобильные номера, существующие в обоих фреймах данных
fines_2 = pd.merge(left=fines, right=owners, on='CarNumber', how='outer')
fines_2
```

```python
#новый фрейм данных должен содержать только номера автомобилей из фрейма данных штрафов
fines_3=pd.merge(left=fines, right=owners, on='CarNumber',how='left')
fines_3
```

```python
#новый фрейм данных должен содержать только номера автомобилей из фрейма данных владельцев
fines_4 = pd.merge(left=fines, right=owners, on='CarNumber', how='right')
fines_4
```

<h2>Создайте сводную таблицу из фрейма данных о штрафах, она должна выглядеть так (значения — это суммы штрафов), но со всеми годами </h2>

```python
pd.pivot_table(fines,
               columns='Year',
               values='Fines',
               index=['Mark', 'Model'],
               aggfunc='sum')
```
