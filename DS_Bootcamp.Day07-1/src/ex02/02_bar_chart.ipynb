```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('../dataset/checking-logs.sqlite')
conn
```

<h2>Анализируйте только пользователей, а не администраторов</h2>

```python
query = """
SELECT 
    timestamp as date
FROM
    checker
WHERE
    uid LIKE 'user_%'
"""
commits = pd.io.sql.read_sql(query, conn, parse_dates=['date'])
commits
```

```python
commits['time'] = pd.cut(commits['date'].dt.hour, bins = [-1, 4,10,17,24], labels=['night','morning','afternoon','evening'])
commits['date']=commits['date'].dt.date
commits
```

```python
commits_group = commits.groupby('time').date.value_counts().unstack(0)
commits_group
```

```python
plt = commits_group.plot.bar(figsize=(15, 8),
                                         fontsize=8,
                                         title='Commits per day',
                                         rot=90,
                                         stacked=True,
                                         color=('red', 'orange', 'c', 'y'))
plt.grid(axis='y')
plt
```

<h2>«Когда наши пользователи обычно выполняют лабораторные работы: ночью, утром, днём или вечером?» — это два наиболее распространённых периода.</h2>

```python
commits_group.sum()
```

Исходя из этого мы можем сделать вывод что наши пользователи выполняют лабароторные работы чаще **днем** и **вечером**.

<h2>Который день имеет:</h2>

- наибольшее количество коммитов

- и в то же время количество коммитов вечером выше, чем днём

```python
print(f"Ответ: {commits_group[commits_group['evening'] > commits_group['afternoon']].sum(axis=1).idxmax()}")
```
