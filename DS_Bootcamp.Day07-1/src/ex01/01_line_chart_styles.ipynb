```python
import pandas as pd
import sqlite3
```

<h2>Соединение с БД</h2>

```python
conn = sqlite3.connect('../dataset/checking-logs.sqlite')
conn
```

<h2>Анализируйте только пользователей, а не администраторов</h2>

```python
query = """
SELECT *
FROM
    pageviews
WHERE uid LIKE 'user_%'
"""

pageview = pd.io.sql.read_sql(query, conn, parse_dates='datetime')
pageview
```

```python
pageview = pageview.groupby(by = pageview['datetime'].dt.date).agg({'uid':['count']})
pageview = pageview.reset_index()
pageview.columns= ['date', 'views']
pageview
```

```python
query = """
SELECT *
FROM 
    checker
WHERE uid LIKE 'user_%'
"""

checker = pd.io.sql.read_sql(query, conn, parse_dates='timestamp')
checker
```

```python
checker = checker.groupby(by = checker['timestamp'].dt.date).agg({'uid': ['count']})
checker = checker.reset_index()
checker.columns = ['date', 'commits']
checker
```

<h2>Анализируйте только даты, когда были как просмотры, так и проверки</h2>

```python
df = pd.merge(pageview, checker, on='date',how='inner')
df
```

<h2>Построение общего графика</h2>

```python
plt = df.plot(x='date',
              y=['views', 'commits'],
              figsize=(15,8),
             fontsize=8,
             title='Views and commits per day',
             rot=90,
             style=['-.', '--'],
             color=['r','c'])
plt.get_lines()[0].set_marker('o')
plt.grid(axis='x')
```
