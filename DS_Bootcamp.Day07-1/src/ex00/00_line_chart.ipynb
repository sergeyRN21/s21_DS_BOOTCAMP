```python
import pandas as pd
import sqlite3
```

<h2>Соединение к бд</h2>

```python
conn = sqlite3.connect('../dataset/checking-logs.sqlite')
conn
```

<h2>Выполните запрос, который получает дату и время из таблицы просмотров страниц, выбирая только пользователей, а не администраторов</h2>

```python
query = """
SELECT *
FROM
    pageviews
WHERE
    uid LIKE 'user_%'
"""
df = pd.io.sql.read_sql(query, conn, parse_dates='datetime')
df
```

<h2>С помощью Pandas создайте новый фрейм данных, в котором подсчитываются посещения и группируются по датам</h2>

```python
df_view = df.groupby(by=df['datetime'].dt.date).agg({'uid':['count']})
df_view = df_view.reset_index()
df_view.columns = ['date', 'view']
df_view
```

<h2>Используя метод Pandas .plot(), создайте график</h2>

- размер шрифта должен быть 8
- размер рисунка равен (15,8)
- график должен содержать заголовок Просмотры за день
- обратите внимание на вращение xticks на графике ниже

```python
df_view.plot(x='date', y='view', figsize=(15,8), title='Views per day', rot=90, fontsize=8)
```
